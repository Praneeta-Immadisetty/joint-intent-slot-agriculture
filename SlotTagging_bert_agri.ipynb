```python
sentence="Which fertilizer makes plants grow faster"
dict={"B-agrochemical_type":["ferti", "pest", "insect", "fungi", "herbicide", "cide"],
"B-agrochemical_price":["cost", "price", "buy", "purchase", "investment", "value", "amount", "charge", "cheap", "expens"],
"B-agrochemical_usecase":["grow", "acid", "use", "usage", "crop", "control", "natural", "artificial", "organic", "protect", "work", "weed", "apply"],
"B-agrochemical_dosage":["dose", "dosage", "quantity", "amount", "litre"],
"B-agrochemical_ingredients":["npk", "zinc", "nitrogen", "potas", "phos", "mix", "variety", "combi", "ingredient", "element", "chemical", "component", "compound", "material", "formula"],
"B-agrochemical_shelflife":["life", "expiry", "duration", "date", "time", "last"],
"B-agrochemical_adv":["benefit", "advantage", "help", "good"],
"B-agrochemical_disadv":["harm", "disadvantage", "effect", "bad"]}
words_list=sentence.split(" ")
print(words_list)
slot_tag=""
for i in words_list:
  keys = [k for k, v in dict.items() for x in v if x in i]
print(keys)
```

```python
l = []
for word in words_list:
    for k in dict:
        #print(k)
        for w in dict[k]:
            if w in word:
                l.append(k)
                break
    l.append('O')
print(l)
```
