```python
def slot(sentence):
  words_list=sentence.split(" ")
  slot_tag=""
  for i in words_list:
    keys = [k for k, v in slots.items() for x in v if x in i]
    if len(keys)!=0:
      slot_tag+=keys[0]+" "
    else:
      slot_tag+="O "
  print("sentence:",sentence,"slot tag:" ,slot_tag)
  return slot_tag
```
